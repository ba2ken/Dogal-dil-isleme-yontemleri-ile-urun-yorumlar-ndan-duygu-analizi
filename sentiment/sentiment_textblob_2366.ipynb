```python
from textblob import TextBlob
```

```python
cevrilmis_df = pd.read_csv("cevrilmis_yorumlar-2366.csv")
```

```python
cevrilmis_list = cevrilmis_df.values.tolist()
```

```python
for i in range(len(cevrilmis_list)):
  listToStr = ' '.join([str(elem) for elem in cevrilmis_list[i]])
  cevrilmis.append(listToStr)
```

```python
for i in range(len(cevrilmis)):
  res = TextBlob(cevrilmis[i])
  skor = res.sentiment.polarity
  print(skor)
  textblob_list.append(skor)
  print(cevrilmis[i])
  if skor < 0:
    print("Negatif")
  elif skor > 0:
    print("Pozitif")
  else:
    print("Nötr")
```

```python
df = pd.read_csv("cevrilmis_yorumlar-2366.csv")
```

```python
df.rename(columns={"Unnamed: 0": "Sıra", "0": "Yorum" } , inplace = True)
```

```python
df.head()
```

```python
df['skor'] = textblob_list
```

```python
df.head()
```

```python
def skor(deger):
  if deger < 0:
    return "Negatif"
  elif deger > 0:
    return "Pozitif"
  else:
    return "Nötr"
```

```python
df['sentiment'] = df['skor'].apply(lambda x: skor(x))
df.head()
```
