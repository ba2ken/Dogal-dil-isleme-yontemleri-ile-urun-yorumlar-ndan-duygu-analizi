```python
from googletrans import Translator
```

```python
cevirici = Translator()
```

```python
cevrilmemis_df = pd.read_csv("duzeltilmis_yorumlar-2366.csv")
```

```python
cevrilmemis_list = cevrilmemis_df.values.tolist()
```

```python
for i in range(len(cevrilmemis_list)):
  listToStr = ' '.join([str(elem) for elem in cevrilmemis_list[i]])
  cevrilmemis.append(listToStr)
```

```python
for i in range (len(cevrilmemis_list)):
  cevrilmis = cevirici.translate(cevrilmemis[i], src='tr', dest='en').text
  cevrilmis_list.append(cevrilmis)
  #print(cevrilmis.text)
```

```python
cevrilmis_df = pd.DataFrame(cevrilmis_list)
```

```python
cevrilmis_df.to_csv("cevrilmis_yorumlar-2366.csv")
```
