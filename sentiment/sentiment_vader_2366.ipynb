```python
df = pd.read_csv('cevrilmis_yorumlar-2366.csv')
```

```python
df.rename(columns={"Unnamed: 0": "Sƒ±ra", "0": "Yorum" } , inplace = True)
```

```python
df['skor'] = df['Yorum'].apply(lambda vader_list: vader.polarity_scores(vader_list))
```

```python
df.head()
```

```python
df['Negatif'] = df['skor'].apply(lambda txt: txt['neg'])
df['N√∂tr'] = df['skor'].apply(lambda txt: txt['neu'])
df['Pozitif'] = df['skor'].apply(lambda txt: txt['pos'])
df['Sonuc'] = df['skor'].apply(lambda txt: txt['compound'])
```

```python
df.head()
```

```python
def polarity_score(compound):
  if compound > 0.05:
    return "Pozitif"
  elif compound < -0.5:
    return "Negatif"
  elif compound >= -0.05 and compound < 0.05:
    return "N√∂tr"
```

```python
df['Sentiment'] = df['Sonuc'].apply(lambda val: polarity_score(val))
df.head()
```

```python
df['Sentiment'].value_counts()
```
