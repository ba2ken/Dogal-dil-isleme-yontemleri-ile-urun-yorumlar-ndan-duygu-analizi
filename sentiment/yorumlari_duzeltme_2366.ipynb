```python
!pip install zemberek_python
import time
import logging
import pandas as pd
ornekler = []
import csv
duzeltilmis_yorumlar = []

from zemberek import (
    TurkishSpellChecker,
    TurkishSentenceNormalizer,
    TurkishSentenceExtractor,
    TurkishMorphology,
    TurkishTokenizer
)
```

```python
logger = logging.getLogger(__name__)
```

```python
yorumlar = pd.read_csv("yorumlar-2366.csv")
yorumlar
type(yorumlar)
```

```python
yorumlar_liste = yorumlar.values.tolist()
len(yorumlar_liste)
#print(yorumlar_liste)
for i in range(len(yorumlar_liste)):
  listToStr = ' '.join([str(elem) for elem in yorumlar_liste[i]])
  ornekler.append(listToStr)

print(ornekler[0])
```

```python
morphology = TurkishMorphology.create_with_defaults()
```

```python
# SENTENCE NORMALIZATION
start = time.time()
normalizer = TurkishSentenceNormalizer(morphology)
logger.info(f"Normalization instance created in: {time.time() - start} s")
```

```python
start = time.time()
for ornek in ornekler:
    #print(ornek)
    #print(normalizer.normalize(ornek), "\n")
    duzeltilmis_yorumlar.append((normalizer.normalize(ornek)))
logger.info(f"Sentences normalized in: {time.time() - start} s")

```

```python
df = pd.DataFrame(duzeltilmis_yorumlar)
```

```python
df.to_csv("duzeltilmis_yorumlar-2366.csv" , index=False)
```
